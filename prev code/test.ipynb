```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.actions.action_builder import ActionBuilder
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from datetime import datetime, timedelta

# 웹드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.instagram.com")

# 로그인 id, password
username = 'zetazetagav@gmail.com'
password = 'dlfjswjfjs!!!'


# 로그인하기
time.sleep(5)
input_id = driver.find_element(By.NAME, 'username')
input_pw = driver.find_element(By.NAME, 'password')
input_id.send_keys(username)
input_pw.send_keys(password)
input_pw.send_keys(Keys.RETURN) #엔터키 입력

time.sleep(10)


# 원하는 사용자 아이디 검색 , 추후 input창으로 수정
search_username = 'seoha1203'
driver.get(f'https://www.instagram.com/{search_username}/')
time.sleep(10)

'''
# 최근 3개월 게시글 링크 가져오기
# 기간 설정
time_gap = datetime.now() - timedelta(days=100)

# 최근 게시글 링크
posts = driver.find_elements(By.XPATH, '//a[contains(@href, "/p/")]')

recent_posts = []

# 최근 게시글 링크 저장
post_links = [post.get_attribute('href') for post in posts]


for post_url in post_links:
    driver.get(post_url)

    try:
        # 날짜 추출 대기
        post_date_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//time'))
        )
        post_date = post_date_element.get_attribute('datetime')
        post_date = datetime.fromisoformat(post_date[:-1])  # 'YYYY-MM-DDTHH:MM:SSZ' 형식 변환

        if post_date >= time_gap:
            recent_posts.append(post_url)

        if len(recent_posts) >= 10:  # 원하는 게시글 수 (여기서는 10개로 설정)
            break

    except Exception as e:
        print(f"게시글 날짜를 가져오는 데 오류 발생: {e}")


if recent_posts:
    print("최근 게시글 링크:")
    for post in recent_posts:
        print(post)
else:
    print("최근 3개월 이내의 게시글이 없습니다.")
'''

# 최근 게시글 10개 가지고 오기
posts = driver.find_elements(By.XPATH, '//a[contains(@href, "/p/")]')

recent_posts = []
post_links = [post.get_attribute('href') for post in posts]  # 링크를 미리 저장

for post_url in post_links:
    driver.get(post_url)

    try:
        recent_posts.append(post_url)  # 모든 게시글 URL을 추가

        # 좋아요 페이지
        likes_element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, '//a[contains(@href, "/liked_by/")]'))
        )

        # 좋아요 누른 사람들 리스트 가져오기
        likes_element.click()  # 좋아요 누른 사람들의 목록 열기
        time.sleep(2)  # 목록이 열릴 때까지 대기
        
        # 좋아요 사용자 리스트 가져오기
        likes_popup = WebDriverWait(driver, 20).until(
                    EC.presence_of_element_located((By.XPATH, '//div[@role="dialog"]//ul'))
        )
        likes = likes_popup.find_elements(By.TAG_NAME, 'li')

        like_users = []
        while True:
            likes = likes_popup.find_elements(By.TAG_NAME, 'li')
            for like in likes:
                like_users.append(like.text)

            # 다음 페이지가 있을 경우 스크롤 내리기
            try:
                next_button = driver.find_element(By.XPATH, '//button[contains(text(), "다음")]')
                next_button.click()
                time.sleep(10)  # 다음 페이지 로딩 대기
            except Exception:
                break  # 다음 버튼이 없으면 종료e_users:
        print(f"\n좋아요한 사용자 (게시글 링크: {post_url}):")
        for user in like_users:
            print(user)
    except Exception as e:
        print(f"좋아요 사용자 리스트를 찾는 데 실패했습니다: {e}")

        if len(recent_posts) >= 10:  # 원하는 게시글 수 (10개로 설정)
            break
    
    # 댓글 찾기
    comments = driver.find_elements(By.XPATH, '//ul[@class="Mr508"]/li')

    comment_users = []
    for comment in comments:
        username = comment.find_element(By.XPATH, './/a').text
        comment_users.append(username)

    print("댓글 단 사용자:")
    for user in comment_users:
        print(user)

    # 게시글 탐색 후 돌아가기
    driver.back()
    time.sleep(3)
    break


# 결과 출력
if recent_posts:
    print("최근 게시글 링크:")
    for post in recent_posts[:10]:  # 10개 이하인 경우 잘라서 출력
        print(post)
else:
    print("최근 게시글이 없습니다.")
```
